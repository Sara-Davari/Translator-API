```python
import speech_recognition as sr
from deep_translator import GoogleTranslator
```

```python
rec = sr.Recognizer()
ex = ('English = en , Spanish = es , Italian = it , French = fr , German = de , Persian = fa , Russian = ru , Japanese = ja , Chinese = zh-CN. ')
lan1 = input('Please enter the language you want your sentence translated to: ' + ex)
time1 = input('Please enter how long your sentence is in seconds: ')
def stt(lan , time):
    with sr.Microphone() as source:
        print('Recording...')
        audio = rec.record(source , duration = int(time))
        print('Processing...')
    text = rec.recognize_google(audio , language = 'auto')
    print("" + text)
    en = GoogleTranslator(source = 'auto' , target = lan).translate(text)
    print(en)
    print('Sentence translated to: ' + lan)
stt(lan1 , time1)
for q in range(0 , 4):
    q = input(str('Would you like to translate another sentence as well?')).strip().lower()
    if q == 'yes':
        print('--------------------------------')
        lan2 = input('Please enter the language you want your sentence translated to: ' + ex)
        time2 = input('Please enter how long your sentence is in seconds: ')
        stt(lan2 , time2)
    else:
        print('See you later.')
        break
```
